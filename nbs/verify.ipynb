```python
import requests
path = "/api/events/"
base_url = "http://localhost:8002"
endpoint = f"{base_url}{path}"
endpoint
```

```python
data = {"page":"/test+"}
```

```python
response = requests.post(endpoint, data=data)
print(response.headers)
print('ok',response.ok)

if response.ok:
  data = response.json()
  print(data,type(data),data.get('items'))
else:
  print(response.text)
```
